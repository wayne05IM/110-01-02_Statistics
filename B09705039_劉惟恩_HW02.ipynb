method: We use bar chart to display this information. <br>

```python
#autolabel function for bar chart
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        plt.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
```

<h3>a.

```python
#read file
df_3_85 = pd.read_excel('Xr03-85.xlsx')
labels = df_3_85['Month']
sizes = df_3_85['Rate']

fig, ax = plt.subplots()
rects = ax.bar(labels, sizes, width=0.3, bottom=None, align='center')

plt.xlabel('Month')
plt.ylabel('Unemployment rate (percent)')
plt.title('Bar Chart of Unemployment rate')
plt.xticks(labels)
plt.ylim(6.0, 10.0)

autolabel(rects)
plt.show()
```

<h3>b.

```python
def geomean(rate):
    rate_1 = rate + 1
    geo_m = math.exp(np.log(rate_1).mean()) - 1
    return geo_m

avg_rate = geomean(df_4_9['Rate of Return'])
print("geometric mean:", avg_rate)
```

<h3>c.

The geometric mean. Since we wanted to calculate the rate of change over the five years, geometric mean is the more reasonable statistic.
