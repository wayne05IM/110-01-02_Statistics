<h2>Problem 2.15

method: We use bar chart to display this information. <br>
reason: We are primarily interested in the numbers. There is no need to present proportions here.

```python
#載入所需函示庫
import matplotlib.pyplot as plt
# 設定圖形大小; DPI越大圖越大
plt.rcParams["figure.dpi"] = 100
import pandas as pd
import numpy as np
```

```python
#autolabel function for bar chart
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        plt.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
```

```python
#read file
df_2_15 = pd.read_excel('Xr02-15.xlsx')
labels = df_2_15['Country']
sizes = df_2_15['Consumption of oil (barrels per day)']

new_sizes = []
for number in sizes:
    new_sizes.append(number / 1000000)
#print(new_sizes)
```

```python
fig, ax = plt.subplots()
rects = ax.bar(labels, new_sizes, width=0.3, bottom=None, align='center')

plt.xlabel('Country')
plt.ylabel('Consumption of Oil (Million Barrels per Day)')
plt.title('Average Oil Consumption per day of the top 20 Oil-consuming Nations')
plt.xticks(labels, rotation = 90)
plt.ylim(0.0, 20.5)
#plt.ylim(0, 20000000)

autolabel(rects)
plt.show()
```

<h2>Problem 2.31

<h3>(a)

method: We use bar chart to display this information. <br>
reason: We are primarily interested in the frequencies. There is no need to present proportions here.

```python
#read file
df_2_31 = pd.read_excel('Xr02-31.xlsx')
sizes = df_2_31['Computer']

labels = ['HP', 'Lenovo', 'Dell', 'Other']
count = [0] * 4
for number in sizes:
    count[number - 1] += 1
    
#bcount = df_2_31['Computer'].value_counts()  #same as the loop above
```

```python
fig, ax = plt.subplots()
rects = ax.bar(labels, count, width=0.5, bottom=None, align='center')

plt.xlabel('Computer Brands')
plt.ylabel('Frequency')
plt.title('Computer Brands Purchased (Frequency)')
plt.xticks(labels)
plt.ylim(0, 45)

autolabel(rects)
plt.show()
```

<h3>(b)

method: We use pie chart to display this information. <br>

```python
#explode = (0.1, 0.1, 0, 0, 0)  # only "explode" the 1st slice (i.e. 'Accounting')
fig1, ax1 = plt.subplots()

ax1.pie(count, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90) #, explode=explode
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Computer Brands Purchased (Porportions)')
plt.show()
```

<h2>Problem 2.43

method: We use complex bar chart to display this information. <br>
reason: We are primarily interested in the differences in types of robberies over the 19-year span. There is no need to present proportions here.

```python
#read file
df_2_43 = pd.read_excel('Xr02-43.xlsx')
count_1995 = df_2_43["Year 1995"].value_counts()
count_2000 = df_2_43["Year 2000"].value_counts()
count_2005 = df_2_43["Year 2005"].value_counts()
count_2009 = df_2_43["Year 2009"].value_counts()

#print(count_1995)
#print(count_1995[1.0])
```

```python
Y_label = ("1995", "2000", "2005", "2009")
x = np.arange(len(Y_label))  # the label locations
width = 0.1  # the width of the bars
fig, ax = plt.subplots()
rects1 = ax.bar(x - 6*width/2, [count_1995[1.0], count_2000[1.0], count_2005[1.0], count_2009[1.0]], width, label = "Street or highway")
rects2 = ax.bar(x - 4*width/2, [count_1995[2.0], count_2000[2.0], count_2005[2.0], count_2009[2.0]], width, label = "Commercial house")
rects3 = ax.bar(x - 2*width/2, [count_1995[3.0], count_2000[3.0], count_2005[3.0], count_2009[3.0]], width, label = "Gas station")
rects4 = ax.bar(x, [count_1995[4.0], count_2000[4.0], count_2005[4.0], count_2009[4.0]], width, label = "Convenience store")
rects5 = ax.bar(x + 2*width/2, [count_1995[5.0], count_2000[5.0], count_2005[5.0], count_2009[5.0]], width, label = "Residence")
rects6 = ax.bar(x + 4*width/2, [count_1995[6.0], count_2000[6.0], count_2005[6.0], count_2009[6.0]], width, label = "Bank")
rects7 = ax.bar(x + 6*width/2, [count_1995[7.0], count_2000[7.0], count_2005[7.0], count_2009[7.0]], width, label = "Other")

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Frequency')
ax.set_xlabel('Year')
ax.set_title('Frequency by the Types of Robberies over the 19-year Span')
ax.set_xticks(x)
ax.set_xticklabels(Y_label)
ax.legend()
plt.ylim(0, 350)
autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
autolabel(rects4)
autolabel(rects5)
autolabel(rects6)
autolabel(rects7)

plt.show()
```

<h2>Problem 2.53

method: We create a contingency table first, then present the contingency table by complex bar chart. <br>
reason: We are primarily interested in the difference between the three political parties. There is no need to present proportions here.

```python
#讀取資料集
df_2_39 = pd.read_excel('Xr02-39.xlsx')
#列出Contingency Table
contb_1 = pd.crosstab(df_2_39["Party"], df_2_39["Social views"])
print("Contingency Table:")
display(contb_1)
```

```python
SV_label = ("Liberal", "Moderate", "Conservative")
P_label = ("Democrat", "Independent", "Republican")
x = np.arange(len(SV_label))  # the label locations
width = 0.2  # the width of the bars
fig, ax = plt.subplots()
rects1 = ax.bar(x - 2*width/2, contb_1[1], width, label = "Liberal")
rects2 = ax.bar(x, contb_1[2], width, label = "Moderate")
rects3 = ax.bar(x + 2*width/2, contb_1[3], width, label = "Conservative")
# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Frequency')
ax.set_xlabel('Party')
ax.set_title('Frequency by Party and Social Issues')
ax.set_xticks(x)
ax.set_xticklabels(P_label)
ax.legend()
plt.ylim(0, 250)
autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
plt.show()
```
